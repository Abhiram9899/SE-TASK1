```python

import matplotlib.pyplot as plt
import numpy as np

def quadratic_model(time, a, b, c):
    temperature = a * (time ** 2) + b * time + c
    return temperature

def main():
    time_values = np.linspace(0, 10, 50)

    try:
        with open('parameters.txt', 'r') as file:
            for line in file:
                try:
                    # Split the line and convert to float
                    a, b, c = map(float, line.strip().split())

                    # Calculate temperatures
                    temperature_calculated = quadratic_model(time_values, a, b, c)

                    # Plot each set of parameters
                    plt.plot(time_values, temperature_calculated, label=f'a={a}, b={b}, c={c}')
                except ValueError as e:
                    print(f"Skipping line due to error: {e}")

    except FileNotFoundError:
        print("The file 'parameters.txt' was not found.")
        return

    plt.xlabel('Time')
    plt.ylabel('Temperature')
    plt.legend()
    plt.title('Weather Modelling with Quadratic Equation Multiple (File Input)')
    plt.show()

if __name__ == "__main__":
    main()
```

```python

import matplotlib.pyplot as plt
import numpy as np

def quadratic_model(time, a, b, c):
    temperature = a * (time ** 2) + b * time + c
    return temperature

def main():
    time_values = np.linspace(0, 10, 50)

    # Define multiple sets of parameters
    parameters_sets = [
        {'a': 1.0, 'b': 2.0, 'c': 3.0, 'label': 'set1'},
        {'a': 0.5, 'b': 1.0, 'c': 2.0, 'label': 'set2'},
        {'a': 2.0, 'b': -1.0, 'c': 5.0, 'label': 'set3'}
    ]

    # Plot each set of parameters
    for params in parameters_sets:
        a = params['a']
        b = params['b']
        c = params['c']
        label = params['label']

        # Calculate temperatures
        temperature_calculated = quadratic_model(time_values, a, b, c)

        # Plot the calculated temperatures
        plt.plot(time_values, temperature_calculated, label=f'{label} (a={a}, b={b}, c={c})')

    plt.xlabel('Time')
    plt.ylabel('Temperature')
    plt.legend()
    plt.title('Weather Modelling with Quadratic Equation (Multiple Lines)')
    plt.show()

if __name__ == "__main__":
    main()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def quadratic_model(time, a, b, c):
    temperature = a * (time ** 2) + b * time + c
    return temperature

def main():
    time_values = np.linspace(0, 10, 50)

    # Define the hardcoded set of parameters
    hardcoded_params = {'a': 0.2, 'b': -5, 'c': 50, 'label': 'Hardcoded Coefficients'}

    # User-provided coefficients
    a_user = float(input("Enter the value of a: "))
    b_user = float(input("Enter the value of b: "))
    c_user = float(input("Enter the value of c: "))
    user_params = {'a': a_user, 'b': b_user, 'c': c_user, 'label': 'User-provided Coefficients'}

    # Combine both sets into a list
    parameters_sets = [hardcoded_params, user_params]

    # Plot each set of parameters
    for params in parameters_sets:
        a = params['a']
        b = params['b']
        c = params['c']
        label = params['label']

        # Calculate temperatures
        temperature_calculated = quadratic_model(time_values, a, b, c)

        # Plot the calculated temperatures
        plt.plot(time_values, temperature_calculated, label=f'{label} (a={a}, b={b}, c={c})')

    plt.xlabel('Time')
    plt.ylabel('Temperature')
    plt.legend()
    plt.title('Weather Modelling with Quadratic Equation')
    plt.show()

if __name__ == "__main__":
    main()
```
